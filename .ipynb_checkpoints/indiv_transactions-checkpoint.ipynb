```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Load Data

```python
delimiter = "|"

transactions = pd.read_csv('itcont_2022_20211214_20220103.txt', sep=delimiter, header=None)
transactions_header = pd.read_csv('indiv_header_file.csv', sep=delimiter, header=None).iloc[0].str.split(",").to_list()[0]

transactions.columns = transactions_header
```

```python
transactions.head(3)
```

```python
cand_commit =  pd.read_csv('ccl.txt', sep=delimiter, header=None)
cand_commit_header = pd.read_csv('ccl_header_file.csv', sep=delimiter, header=None).iloc[0].str.split(",").to_list()[0]

cand_commit.columns = cand_commit_header
```

```python
cand_commit.head(3)
```

```python
cand_name =  pd.read_csv('cn.txt', sep=delimiter, header=None)
cand_name_header = pd.read_csv('cn_header_file.csv', sep=delimiter, header=None).iloc[0].str.split(",").to_list()[0]

cand_name.columns = cand_name_header
```

```python
cand_name.head(3)
```

```python
part_incumb = house.groupby(['CAND_PTY_AFFILIATION', 'CAND_ICI']).size().reset_index(name='contributions')

plt.figure(figsize=(10, 6))
bar_plot = plt.bar(x=part_incumb.index, height=part_incumb['contributions'], tick_label=part_incumb['CAND_PTY_AFFILIATION'] + " - " + part_incumb['CAND_ICI'])
plt.xlabel('Party - Incumbent Status')
plt.ylabel('Number of Contributions')
plt.title('Number of Individual Campaign Contributions by Party and Incumbent Status')
plt.xticks(rotation=45)
plt.show()
```

```python
bins = np.arange(0, 2000, 100)
plt.figure(figsize=[9, 6])
sns.histplot(incumbent_contributions[incumbent_contributions.num_contributions < 2000], x='num_contributions', stat='density', bins=bins, palette='tab10', alpha=0.5, kde=True, color='blue', label='Incumbent')
sns.histplot(open_contributions[open_contributions.num_contributions < 2000], x='num_contributions', stat='density', bins=bins, palette='tab10', alpha=0.5, kde=True, color='red', label='Non-incumbent')
plt.legend()
plt.title('Distributions of # of individual contributions among primaries involving incumbents vs. primaries without incumbents')
plt.xlabel('# of contributions');
```
